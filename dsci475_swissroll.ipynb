# Isomap Dimensionality Reduction on Swiss Roll Data

```python
#import libraries
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn import manifold, datasets
```

```python
#create Swiss Roll
X, color = datasets.make_swiss_roll(n_samples=3000, random_state=0)

#plot 3D Swiss Roll
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection = "3d")
fig.add_axes(ax)
ax.scatter(
    X[:, 0], X[:, 1], X[:, 2], c = color, s = 10, alpha = 0.8
)
ax.set_title("Swiss Roll in 3D")
ax.view_init(azim = -66, elev = 12)
_ = ax.text2D(0.8, 0.05, s="n = 3,000", transform = ax.transAxes)
```

```python
# run Isomap on the points in X with 2 dim output
Y = manifold.Isomap(n_neighbors = 10, n_components = 2).fit_transform(X)

# 2D projection
plt.figure()
scatter(Y[:,0], Y[:,1], c = color, s = 30, alpha = 0.8)
title("Swiss Roll in 2D")
figtext(0.8, 0.05, s = "n = 3,000")
show()
```

```python
#Swiss Roll with hole
X_hole, color_hole = datasets.make_swiss_roll(n_samples = 3000, hole = True, random_state = 0)

fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection = "3d")
fig.add_axes(ax)
ax.scatter(
    X_hole[:, 0], X_hole[:, 1], X_hole[:, 2], c = color_hole, s = 10, alpha = 0.8
)
ax.set_title("Swiss Role with Hole in 3D")
ax.view_init(azim = -66, elev = 12)
_ = ax.text2D(0.8, 0.05, s = "n_samples = 3,000", transform = ax.transAxes)
```

```python
# run Isomap on the points in X with 2 dim output
Y_hole = manifold.Isomap(n_neighbors = 10, n_components = 2).fit_transform(X_hole)

# 2D projection
plt.figure()
scatter(Y_hole[:,0], Y_hole[:,1], c = color_hole, s = 30, alpha = 0.7)
title("Swiss Roll in 2D")
figtext(0.8, 0.05, s = "n = 3,000")
show()
```
